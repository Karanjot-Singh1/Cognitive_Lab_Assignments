a) Reverse the NumPy array: arr = np.array([1, 2, 3, 6, 4, 5])

```python
import numpy as np;
arr=np.array([1,2,3,6,4,5])
print("print the original array :")
for i in range(6):
  print(arr[i],end=" ")
print()

j=5
for i in range(6//2):
  temp=arr[i]
  arr[i]=arr[j]
  arr[j]=temp
  j=j-1

print("the reverse array is : ")
for i in range(6):
  print(arr[i],end=" ")
```
