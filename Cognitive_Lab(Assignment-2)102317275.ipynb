**1. Create a List L that is defined as= [10, 20, 30, 40, 50, 60, 70, 80].  i. WAP to add 200 and 300 to L.  
ii. WAP to remove 10 and 30 from L.  
iii. WAP to sort L in ascending order.  
iv. WAP to sort L in descending order.**  


```python
L=[10,20,30,40,50,60,70,80]
L.append(200)
L.append(300)
print("the list after adding 200 and 300 is : ")
print(L)

L.remove(10)
L.remove(30)
print("the list after removing 10 and 30 is : ")
print(L)

L.sort()
print("the list after sorting in ascending order : ")
print(L)

L.sort(reverse=True)
print("the list after sorting in decending order : ")
print(L)
```

iii. Reverse the tuple and return it as a list.

**3. WAP to create a list of 100 random numbers between 100 and 900. Count and print  the:  
i. All odd numbers  
ii. All even numbers  
iii. All prime numbers**


```python
import random as r
l=[]
for i in range(100):
  l.append(r.randint(100,900))
print("the list is: ")
print (l)

print("\nprinting all odd numbers: ",)
for num in l:
  if(num%2==1):
    print(num,end=" ")

print("\n\nprinting all even numbers: ")
for num in l:
  if(num%2==0):
    print(num,end=" ")

print("\n\nprinting all prime numbers: ")
for num in l:
  f=0
  for i in range(2,(num//2)+1):
    if(num%i==0):
      f=1
      break
  if f==0:
    print(num,end=" ")

```

**4. Consider the following two sets, A and B, representing scores of two teams in multiple  matches. A = {34, 56, 78, 90} and B ={78, 45, 90, 23} WAP to perform the following operations using set functions:
i.) Find the unique scores achieved by both teams (union of sets)
ii.) Identify the scores that are common to both teams (intersection of sets).
iii.) Find the scores that are exclusive to each team (symmetric difference).
iv.) Check if the scores of team A are a subset of team B, and if team B's scores are  a superset of team A.  
v.) Remove a specific score __ (input by the user) from set A if it exists. If not, print  a message saying it is not present.**  


```python
A={34,56,78,90}
B={78,45,90,23}
print("the unique scores achieved by both teams",A.union(B))

print("the scores that are common to both teams ",A.intersection(B))

print("the scores that are exclusive to each team",A^B)

if A.issubset(B):
  print("set A is subset of the set B")
else:
  print("set A is not the subset of the set B")
if B.issuperset(A):
  print("set B is superset of the set A")
else:
  print("set B is not the superset of the set A")

searching_ele=int(input("enter the searching element: "))
if searching_ele in A:
  print("the searching ele is present in the set")
else:
  print("the searching ele is not present in the set")

```
